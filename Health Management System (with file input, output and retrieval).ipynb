```python
def getdate():
    import datetime
    return datetime.datetime.now()


def log(k):
    if k == 1:
        c = int(input('Enter 1 for exercise and 2 for food'))
        if c == 1:
            value = input('Enter exercise name: \n')
            with open("Ritik_exercise_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')
        if c == 2:
            value = input('Enter Food details: \n')
            with open("Ritik_food_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')

    if k == 2:
        c = int(input('Enter 1 for exercise and 2 for food'))
        if c == 1:
            value = input('Enter exercise name: \n')
            with open("Manushri_exercise_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')
        if c == 2:
            value = input('Enter Food details: \n')
            with open("Manushri_food_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')

    if k == 3:
        c = int(input('Enter 1 for exercise and 2 for food'))
        if c == 1:
            value = input('Enter exercise name: \n')
            with open("Swastik_exercise_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')
        if c == 2:
            value = input('Enter Food details: \n')
            with open("Swastik_food_data.txt", 'a') as op:
                op.write(str([str(getdate())]) + ": " + value + '\n')
            print('Saved successfully')
        else:
            print("plz enter valid input (1(Ritik),2(Manushri),3(Swastik)")


def retrieve(k):
    if k == 1:
        c = int(input("enter 1 for excersise and 2 for food"))
        if c == 1:
            with open("Ritik_exercise_data.txt") as op:
                for i in op:
                    print(i, end="")
        elif c == 2:
            with open("Ritik_food_data.txt") as op:
                for i in op:
                    print(i, end="")

    if k == 2:
        c = int(input("enter 1 for excersise and 2 for food"))
        if c == 1:
            with open("Manushri_exercise_data.txt") as op:
                for i in op:
                    print(i, end="")
        elif c == 2:
            with open("Manushri_food_data.txt") as op:
                for i in op:
                    print(i, end="")

    if k == 3:
        c = int(input("enter 1 for excersise and 2 for food"))
        if c == 1:
            with open("Swastik_exercise_data.txt") as op:
                for i in op:
                    print(i, end="")
        elif c == 2:
            with open('Swastik_food_data.txt') as op:
                for i in op:
                    print(i, end="")
        else:
            print("plz enter valid input (Ritik, Manushri, Swastik)")
print("health management system: ")
a = int(input("press 1 for logging the value and 2 for retrieve "))

if a != 1:
    b = int(input("press 1 for Ritik 2 for Manushri 3 for Swastik "))
    retrieve(b)
else:
    b = int(input("press 1 for Ritik 2 for Manushri 3 for Swastik "))
    log(b)
```
